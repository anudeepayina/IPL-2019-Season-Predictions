```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
final_2008 = pd.read_csv("final_2008_.csv")
final_2009 = pd.read_csv("final_2009_.csv")
final_2010 = pd.read_csv("final_2010_.csv")
final_2011 = pd.read_csv("final_2011_.csv")
final_2012 = pd.read_csv("final_2012_.csv")
final_2013 = pd.read_csv("final_2013_.csv")
final_2014 = pd.read_csv("final_2014_.csv")
final_2015 = pd.read_csv("final_2015_.csv")
final_2016 = pd.read_csv("final_2016_.csv")
final_2017 = pd.read_csv("final_2017_.csv")
final_2018 = pd.read_csv("final_2018_.csv")
final_2019 = pd.read_csv("final_2019_.csv")
```

```python
##Make large dataset
final = pd.concat([final_2008,final_2009,final_2010,final_2011,final_2012,final_2013,final_2014,final_2015,final_2016,final_2017,final_2018,final_2019])
```

```python
final = final[['season', 'team1', 'team2', 'winner','net_SRH', 'net_RCB', 'net_MI',
       'net_RPS', 'net_GL', 'net_KKR', 'net_KXIP', 'net_DD', 'net_CSK',
       'net_RR', 'net_DC', 'net_KTK', 'net_PW', 'net_RPS.1', 'net_DCa',
       'Sunrisers Hyderabad', 'Royal Challengers Bangalore', 'Mumbai Indians',
       'Rising Pune Supergiant', 'Gujarat Lions', 'Kolkata Knight Riders',
       'Kings XI Punjab', 'Delhi Daredevils', 'Chennai Super Kings',
       'Rajasthan Royals', 'Deccan Chargers', 'Kochi Tuskers Kerala',
       'Pune Warriors', 'Rising Pune Supergiants', 'Delhi Capitals']]
final.dropna(subset=["winner","team1","team2"],inplace=True)
#final
```

```python
##Use standard scaler on net run rate and winning percentage
scaler = StandardScaler()
mmx = MinMaxScaler()
final.iloc[:,4:] = mmx.fit_transform(final.iloc[:,4:])
final
```

```python
final.loc[final["winner"]==final["team1"],"winning_team"]=1
final.loc[final["winner"]==final["team2"],"winning_team"]=0
final = pd.get_dummies(final,prefix=["team1","team2"],columns=["team1","team2"])
final.drop(["winner"],axis=1,inplace=True)
```

```python
final.fillna(0,inplace=True)
final.head()
```

```python
final.to_csv("finaldata.csv",index=False)
```
